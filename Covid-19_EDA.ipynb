### Data Visualization

#### Univariate analysis

```python
#df.nlargest(10,'Confirmed')
#df[['Country/Region', 'Confirmed']].sort_values(by='Confirmed', ascending=False).head(10)
```

```python
#Top 10 Countries by confirmed cases
#Country vs confirmed cases

top_confirmed = df.sort_values('Confirmed', ascending=False).head(10)
plt.figure(figsize=(10,6))
ax=sns.barplot(y="Country/Region", x="Confirmed", data=top_confirmed, palette='Reds_r')
plt.xlabel("Total Confirmed Cases")
plt.ylabel("Country")
plt.title("Top 10 Countries by confirmed cases")
plt.show()
```

```python
top_recovered = df.sort_values('Recovered', ascending=False).head(10)
plt.figure(figsize=(10,6))
ax=sns.barplot(y="Country/Region", x="Recovered", data=top_recovered, palette='Reds_r')
plt.xlabel("Total Recovered Cases")
plt.ylabel("Country")
plt.title("Top 10 Countries by recovered cases")
plt.show()
```

```python
top_deaths = df.sort_values('Deaths', ascending=False).head(10)
plt.figure(figsize=(10,6))
ax=sns.barplot(y="Country/Region", x="Deaths", data=top_deaths, palette='Reds_r')
plt.xlabel("Total Deaths Cases")
plt.ylabel("Country")
plt.title("Top 10 Countries by Deaths cases")
plt.show()
```

```python
#Histplot:Distribution of Death Rate

plt.figure(figsize=(8, 6))
sns.histplot(df['Deaths / 100 Cases'], bins=30, kde=True,color='red')
plt.title("Distribution of Death Rate")
plt.xlabel("Deaths per 100 Cases (%)")
plt.show()
```

```python
#Histogram : Recovery Rate Distribution

plt.figure(figsize=(8, 6))
ax=sns.histplot(df['Recovered / 100 Cases'], bins=20, kde=True, color='red')
plt.xlabel("Recovery Rate (%)")
plt.ylabel("Frequency")
plt.title("Distribution of Recovery Rates Across Countries")
plt.show()
```

```python
#Countries with Highest % Increase Last Week

# Top 10 countries with the highest weekly % increase
top_growth = df[['Country/Region', '1 week % increase']].sort_values(by='1 week % increase', ascending=False).head(10)

# Bar plot
plt.figure(figsize=(12, 6))
ax=sns.barplot(x='1 week % increase', y='Country/Region', data=top_growth, palette='Reds_r')
for bar in ax.containers:
    ax.bar_label(bar)
plt.xlabel("1 Week % Increase")
plt.ylabel("Country")
plt.title("Top 10 Countries with Highest Weekly Growth")
plt.show()
```

```python
plt.figure(figsize = (10, 6))
ax=sns.barplot(x = 'WHO Region', y = 'Active',palette='Reds_r', data = df)
for bar in ax.containers:
    ax.bar_label(bar)
plt.title('Active Cases by Who Region')
plt.show()
```

#### Bivariate Analysis

```python
#Scatterplot: confirmed Cases Vs Deaths Cases
plt.figure(figsize = (8, 6))
sns.scatterplot(data=df, x='Confirmed', y='Deaths')
plt.title("Confirmed Cases vs Deaths")
plt.grid(True)
plt.show()
```

```python
# Scatter plot: Death Rate vs. Recovery Rate
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['Deaths / 100 Cases'], y=df['Recovered / 100 Cases'], alpha=0.7)
plt.xlabel("Death Rate (%)")
plt.ylabel("Recovery Rate (%)")
plt.title("Death Rate vs. Recovery Rate")
plt.grid(True)
plt.show()
```

```python
# Line plot: Active vs. Recovered cases
plt.figure(figsize=(8,5))
sns.lineplot(x=df['Active'],y=df['Recovered'])
plt.title("Active cases Vs Recovered",fontsize=14)
```

```python
#Mortality Rate Across WHO Regions

# Boxplot of death rate by WHO region
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['WHO Region'], y=df['Deaths / 100 Cases'], palette='Reds_r')
plt.xlabel("WHO Region")
plt.ylabel("Death Rate (%)")
plt.title("Mortality Rate Across WHO Regions")
plt.xticks(rotation=45)
plt.show()
```

#### Multivariate Analysis

```python
#COVID-19 Impact by WHO Region

region_group = df.groupby('WHO Region')[['Confirmed', 'Deaths', 'Recovered', 'Active']].sum().sort_values('Confirmed', ascending=False)

region_group.plot(kind='bar', figsize=(12, 6))
plt.title("COVID-19 Impact by WHO Region")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
#Correlation 
# Select only numerical columns
numeric_df = df.select_dtypes(include=['number'])

# Correlation heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of COVID-19")
plt.xticks(rotation=45)
plt.show()
```

```python
#Countries with zero Active cases

# Countries with zero active cases
zero_active = df[df['Active'] == 0]

# Count plot
plt.figure(figsize=(10, 5))
ax=sns.countplot(x=zero_active['WHO Region'], palette='Reds_r')
for bar in ax.containers:
    ax.bar_label(bar)
plt.xlabel("WHO Region")
plt.ylabel("Count")
plt.title("Countries with Zero Active Cases by WHO Region")
plt.show()
```
